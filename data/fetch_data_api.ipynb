```python
# Import Libraries
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Code to fetch dataset from Kaggle API
# Install dependencies as needed:
# pip install kagglehub[pandas-datasets]
import kagglehub
from kagglehub import KaggleDatasetAdapter

# Set the path to the file you'd like to load
file_path = "financial_fraud_detection_dataset.csv"

# Load the latest version
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "aryan208/financial-transactions-dataset-for-fraud-detection",
  file_path,
  # Provide any additional arguments like 
  # sql_query or pandas_kwargs. See the 
  # documenation for more information:
  # https://github.com/Kaggle/kagglehub/blob/main/README.md#kaggledatasetadapterpandas
)

print("First 5 records:", df.head())
```

Some notes on the sql_query parameter in Kaggle's load_dataset function:

The sql_query parameter allows us to filter and transform data before loading it into memory. This is useful when:
The dataset is very large and we only need specific columns
we want to pre-filter rows
we need to perform joins or aggregations 

```python
# After your KaggleHub fetch

# We save a raw snapshot (keeps our pipeline reproducible)
df.to_csv("data/financial_fraud_detection_dataset.csv", index=False)

# Optional: save as Parquet for compact storage
df.to_parquet("data/financial_fraud_detection_dataset.parquet")
```
