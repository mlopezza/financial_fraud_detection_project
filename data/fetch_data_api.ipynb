# Fraud Detection Scores

## 1. Spending Deviation Score
- **Measures:** How unusual a transaction amount is compared to the customer’s historical spending.  
- **Calculation:** Often based on statistical deviation (e.g., z‑score).  
  

\[
  \text{score} = \frac{\text{current amount} - \mu}{\sigma}
  \]

  
  where \(\mu\) = average spending, \(\sigma\) = standard deviation.  
- **Interpretation:**  
  - High score → transaction amount is far from normal → possible fraud.  
  - Low score → transaction is consistent with past behavior.

---

## 2. Velocity Score
- **Measures:** The speed and frequency of transactions within a short time window.  
- **Calculation:** Count transactions per unit of time and compare with historical averages.  
  Example: customer usually makes 2 transactions/day, but today 15 in 30 minutes.  
- **Interpretation:**  
  - High score → unusually rapid activity → potential card theft or automated fraud.  
  - Low score → normal transaction pace.

---

## 3. Geo Anomaly Score
- **Measures:** Geographic inconsistencies in transaction locations.  
- **Calculation:** Compare current transaction location with previous ones; check distance and time feasibility.  
  Example: purchase in Toronto followed by another in Tokyo within 10 minutes.  
- **Interpretation:**  
  - High score → impossible or improbable travel → suspicious activity.  
  - Low score → location consistent with customer’s usual pattern.

---

##  Summary
- **Spending Deviation Score** → detects abnormal amounts.  
- **Velocity Score** → detects abnormal transaction frequency.  
- **Geo Anomaly Score** → detects abnormal geographic patterns.  
Together, they form part of a broader **fraud risk model** that flags transactions for review or rejection.
